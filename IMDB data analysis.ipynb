```python
# classifying movies based on Ratings (Excellent, Good and Average)
def rate(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Rating Category']=data['Rating'].apply(rate)
```
